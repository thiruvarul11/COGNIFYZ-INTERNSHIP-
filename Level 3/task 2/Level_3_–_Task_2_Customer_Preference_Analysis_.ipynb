```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
df=pd.read_csv('zomato.csv',encoding='latin1')
df.head()
df['Cuisines'].head(10)
df=df.dropna(subset=['Cuisines'])
df['Primary Cuisine']=df['Cuisines'].apply(lambda x: x.split(',')[0].strip())
popular_cuisines=df.groupby('Primary Cuisine')['Votes'].sum().sort_values(ascending=False).head(10)
print(popular_cuisines)
plt.figure(figsize=(10,6))
sns.barplot(x=popular_cuisines.values,y=popular_cuisines.index,palette='Spectral')
plt.title("Top 10 Most Popular Cuisines by Votes")
plt.xlabel("Total Votes")
plt.ylabel("Cuisine")
plt.show()
avg_rating_by_cuisine=df.groupby('Primary Cuisine')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print(avg_rating_by_cuisine)
plt.figure(figsize=(10,6))
sns.barplot(x=avg_rating_by_cuisine.values,y=avg_rating_by_cuisine.index,palette='coolwarm')
plt.title("Top 10 Cuisines with Highest Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Cuisine")
plt.show()
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Aggregate rating',y='Votes',hue='Primary Cuisine',legend=False,alpha=0.5)
plt.title("Rating vs Votes by Primary Cuisine")
plt.xlabel("Aggregate Rating")
plt.ylabel("Votes")
plt.show()
```
