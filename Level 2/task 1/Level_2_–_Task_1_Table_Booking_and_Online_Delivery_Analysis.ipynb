```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns',None)
df=pd.read_csv('zomato.csv',encoding='latin1')
df.head()
print("Table Booking Options:",df['Has Table booking'].unique())
print("Online Delivery Options:",df['Has Online delivery'].unique())
print("Price Ranges:",df['Price range'].unique())
df['Has Table booking']=df['Has Table booking'].str.strip()
df['Has Online delivery']=df['Has Online delivery'].str.strip()
table_booking_counts=df['Has Table booking'].value_counts()
print("Table Booking Count:\n",table_booking_counts)
table_booking_percent=(table_booking_counts / df.shape[0]) * 100
print("Table Booking Percentage:\n",table_booking_percent)
plt.pie(table_booking_counts,labels=table_booking_counts.index,autopct='%1.1f%%',colors=['skyblue','lightgreen'])
plt.title('Table Booking Availability')
plt.show()
online_delivery_counts=df['Has Online delivery'].value_counts()
print("Online Delivery Count:\n",online_delivery_counts)
online_delivery_percent=(online_delivery_counts / df.shape[0]) * 100
print("Online Delivery Percentage:\n",online_delivery_percent)
plt.pie(online_delivery_counts,labels=online_delivery_counts.index,autopct='%1.1f%%',colors=['orange','lightcoral'])
plt.title('Online Delivery Availability')
plt.show()
df['Table_Booking_Binary']=df['Has Table booking'].map({'Yes': 1,'No': 0})
rating_comparison=df.groupby('Table_Booking_Binary')['Aggregate rating'].mean()
print("Average Ratings (0 = No, 1 = Yes):\n",rating_comparison)
sns.barplot(x=['No','Yes'],y=rating_comparison.values, palette='Set2')
plt.title('Average Rating:Table Booking vs No Table Booking')
plt.ylabel('Average Rating')
plt.xlabel('Table Booking Available')
plt.show()
```
